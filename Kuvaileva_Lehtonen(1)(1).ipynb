```python
# avataan data frameen ja tarkistetaan
df1 = pd.read_excel('http://taanila.fi/kunnat.xlsx')
df1.head()
```

```python
# Alempana luokittelua tehdessä, huomasin, että "Väkiluku, 2018" sarakkeen nimen voisi vaihtaa ainakin käsittelyn
# ajaksi lyhyempään. Tai ainakin poistaa pilkku ja välilyönti. Teen sen tähän väliin.
df1 = df1.rename(columns={'Väkiluku, 2018':'Väkiluku2018'})
df1.head()
```

```python
# Tsekataan huvikseen arvojen määrä kaikissa sarakkeissa. Tilastokeskuksen tuntien ja datan luonteen huomioon ottaen,
# jokaisessa sama määrä arvoja
df1.count()
```

```python
df1.isnull()
```

```python
# Katsotaan toisinkin päin
df1.isnull().sum()
```

```python
for var in df1:
    print(var, pd.unique(df1[var]))
```

```python
# Teen kunnan väkiluvuille luokittelun. Tuplaan määrän joka askeleella, jotta saadaan järkevä määrä luokkia. Viimeiseen
# tietenkin tulee vain Helsinki.. Lisään luokille vielä erikseen nimet.

bins = [0, 5000, 10000, 20000, 40000, 80000, 160000, 320000, 640000, 1280000]
names = ['0 - 5000', '5001 - 10000', '10001 - 20000', '20001 - 40000', '40001 - 80000', '80001 - 160000', '160001 - 320000',
        '320001 - 640000', '640000 - 1280000']

df1['Väkiluku_luok'] = pd.cut(df1['Väkiluku2018'], bins = bins, labels = names)
```

```python
# Katsotaan, että homma toimi.
df1.head()
```

```python
# Mielenkiinnosta katsotaan syntyvyys, koska olen Kouvolasta. Sarakkeen nimi on aika hirveä, muutan sen ensin.
df1 = df1.rename(columns={'Syntyneiden enemmyys, henkilöä, 2018':'Syntyvyys2018'})
df1.head()
```

```python
# Kouvolan kunnan kaikki muuttujat.
df1[df1['Kunta']=='Kouvola']
```

```python
# Syntyvyyden mukaan lajiteltuna nähdään että Kouvola pitää viimeistä sijaa.
df1[df1['Syntyvyys2018']<-200].sort_values('Syntyvyys2018')
```

```python
# Haluan ottaa mukaan myös väkiluvun muutoksen, nimeän siis sarakkeen uudestaan jotta helpompi kirjoitella
df1 = df1.rename(columns={'Väkiluvun muutos edellisestä vuodesta, %, 2018':'Väkiluku_muutos%'})
df1.head()
```

```python
# Tässä nopeasti tarkasteltuna syntyvyyden häntäpään kuntia ja lajiteltuna syntyvyyden mukaan.
df1[['Kunta', 'Syntyvyys2018', 'Väkiluku2018', 'Väkiluku_muutos%']][df1['Syntyvyys2018']<-200].sort_values('Syntyvyys2018')
```

```python
# Nimetään taas sarakkeita
df1 = df1.rename(columns={"Taajama-aste, %, 2017":"Taajama%"})
df1.head()
```

```python
# Muutama keskiarvo
print("Väkilukujen keskiarvo:", df1['Väkiluku2018'].mean())
print("Taajama-aste prosentteina, keskiarvo:", df1['Taajama%'].mean())
print("Väkiluvun muutos prosentteina, keskiarvo:", df1['Väkiluku_muutos%'].mean())
```

```python
# Frekvenssitaulukko kuntien väkilukuluokituksen mukaan. Nimeän dataframen df2, jotta tämä ei vaikuta df1-frameen.
df2 = pd.crosstab(df1['Väkiluku_luok'], 'Kuntien lukumäärä')
df2.columns.name = ''

# Prosenttilaskenta
n = df2['Kuntien lukumäärä'].sum()
df2['%'] = (df2['Kuntien lukumäärä']/n)*100

# Tyyppimuunnos ja yhteensä-rivi
df2.index = df2.index.astype(str)
df2.loc['Yhteensä'] = df2.sum()

# Tyylit
df2.style.format({'%':'{:.1f} %'})
```

```python
# Ensimmäinen grafiikka käyttäen df2-framea
ax1 = df2['%'].plot.barh()
ax1.set_xlabel('Prosenttia kaupungeista, n=' + str(n))

ax1.set_xticklabels(['{:.0f} %'.format(x) for x in ax1.get_xticks()])
```

```python
# Syntyvyyden esittäminen boxplotilla. Tästä huomaa, että suurin osa syntyvyydestä on keskittynyt nollan molemmin puolin,
# paino kuitenkin negatiivisella puolella. Kuvaajassa myös paljon poikkeavia lukuja.

# Fontissa täytyi muistaa, että aiempi määritelmä jää voimaan (voidaan määrittää defaulteiksi heti alussa).
# Fontin koko siis pitää määrittää uudestaan sopivaksi.

sns.set(font_scale=1)
ax3 = sns.boxplot(df1['Syntyvyys2018'])
ax3.set_xlabel('Syntyvyys 2018')
```

```python
# Esitetään taajaman osuus kunnan pinta-alasta boxplottina. Poikkeamia ovat kunnat, joilla taajamaosuus on nolla %
# Aiemmasta grafiikasta kopioitu prosenttimerkkimuotoilu
ax4 = sns.boxplot(df1['Taajama%'])
ax4.set_xlabel('Taajaman osuus kunnan pinta-alasta')

ax4.set_xticklabels(['{:.0f} %'.format(x) for x in ax4.get_xticks()])
```

```python
# Nimetään uudelleen sarakkeita. Teen tämän nyt näin, koska tunsin sen itselleni toistaiseksi helpoimmaksi. Aaltosulut
# vähän hämäävät vielä.
df1 = df1.rename(columns={'Alueella olevien työpaikkojen lukumäärä, 2016':'Työpaikat'})
df1 = df1.rename(columns={'Alueella asuvan työllisen työvoiman määrä, 2017':'Työvoima'})
df1 = df1.rename(columns={'Työllisyysaste, %, 2017':'Työllisyysaste%'})
df1 = df1.rename(columns={'Työttömien osuus työvoimasta, %, 2017':'Työttömät%'})
df1.head()
```

```python
# Vähän alkeellista työllisyysseurantaa. Niiltä osin missä on enemmän työvoimaa kuin työpaikkoja, voidaan olettaa että
# pendelöidään lähikuntiin ja isompiin kaupunkeihin. Taulukot on järjestetty työllisyysasteen mukaan.
df4 = df1[['Kunta','Työpaikat', 'Työvoima', 'Työllisyysaste%', 'Työttömät%']].sort_values('Työllisyysaste%', ascending=False)
df4.tail(10)
```
